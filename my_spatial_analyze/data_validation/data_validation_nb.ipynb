```python
hybas_id_list = [
    1020000010, 1020011530, 1020018110, 1020021940, 1020027430, 1020034170, 1020035180, 1020040190,
    2020000010, 2020003440, 2020018240, 2020024230, 2020033490, 2020041390, 2020057170, 2020065840, 2020071190,
    3020000010, 3020003790, 3020005240, 3020008670, 3020009320, 3020024310,
    4020000010, 4020006940, 4020015090, 4020024190, 4020034510, 4020050210, 4020050220, 4020050290, 4020050470,
    5020000010, 5020015660, 5020037270, 5020049720, 5020082270, 
    6020000010, 6020006540, 6020008320, 6020014330, 6020017370, 6020021870, 6020029280,
    7020000010, 7020014250, 7020021430, 7020024600, 7020038340, 7020046750, 7020047840, 7020065090,
    8020000010, 8020008900, 8020010700, 8020020760, 8020022890, 8020032840, 8020044560,
    9020000010
]

from datetime import datetime
from dateutil import relativedelta
import pandas as pd
import data_validation

comparison_metrics_save_paths = []
for basin_id in hybas_id_list:
    print(f'Processing basin {basin_id}')
    my_lakes_path = f'/WORK/Data/global_lake_area/area_csvs/levels_corrected_using_ICESat2/level_{basin_id}_corrected.csv'
    ICESat2_median_path = f'/WORK/Data/global_lake_area/ICESat2_water_levels/time_series_csvs/{basin_id}_ICESat2_time_series.csv'
    ICESat2_min_path = f'/WORK/Data/global_lake_area/ICESat2_water_levels/time_series_csvs_min/{basin_id}_ICESat2_time_series.csv'
    ICESat2_max_path = f'/WORK/Data/global_lake_area/ICESat2_water_levels/time_series_csvs_max/{basin_id}_ICESat2_time_series.csv'
    ICESat2_count_path = f'/WORK/Data/global_lake_area/ICESat2_water_levels/time_series_csvs_monthly_count/{basin_id}_ICESat2_time_series.csv'
    
    ICESat2_df = pd.read_csv(ICESat2_median_path)
    ICESat2_column_names = ICESat2_df.columns.to_list()
    ICESat2_column_names.remove('id')
    
    my_lakes_start_date = '2001-01-01'
    my_lakes_end_date = '2024-01-01'
    my_lakes_start_date = datetime.strptime(my_lakes_start_date, '%Y-%m-%d')
    my_lakes_end_date = datetime.strptime(my_lakes_end_date, '%Y-%m-%d')
    current_date = my_lakes_start_date
    mylakes_columns = []
    while current_date < my_lakes_end_date:
        mylakes_columns.append(current_date.strftime('%Y-%m-%d'))
        current_date += relativedelta.relativedelta(months=1)
    
    save_path = f'/WORK/Data/global_lake_area/validation_data/ICESat2_corrected_with_ICESat2/{basin_id}_comparison_metrics.csv'
    comparison_metrics_save_paths.append(save_path)
    data_validation.compare_corrected_water_levels_with_ICESat2(
        corrected_water_levels_csv_path=my_lakes_path,
        ICESat2_water_levels_monthly_median_csv_path=ICESat2_median_path,
        ICESat2_water_levels_monthly_min_csv_path=ICESat2_min_path,
        ICESat2_water_levels_monthly_max_csv_path=ICESat2_max_path,
        ICESat2_water_levels_monthly_count_csv_path=ICESat2_count_path,
        corrected_water_level_column_names=mylakes_columns,
        ICESat2_water_level_column_names=ICESat2_column_names,
        comparison_metrics_save_path=save_path,
        minimum_valid_points_for_comparison=12
    )

# concatenate all saved comparison metrics
comparison_metrics_dfs = [
    pd.read_csv(save_path) for save_path in comparison_metrics_save_paths
]
concatenated_comparison_metrics = pd.concat(comparison_metrics_dfs)
concatenated_comparison_metrics_save_path = '/WORK/Data/global_lake_area/validation_data/ICESat2_corrected_with_ICESat2/concatenated_comparison_metrics.csv'
concatenated_comparison_metrics.to_csv(concatenated_comparison_metrics_save_path, index=False)
```
